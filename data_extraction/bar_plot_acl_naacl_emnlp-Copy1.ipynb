```python
from bs4 import BeautifulSoup
```

```python
x=[]
y=[]
```

```python
for j in range(79,100,2):
    n1='acl/acl'+str(j)+'.txt'
    #n2='naacl/naacl'+str(j)+'.txt'
    #n3='emnlp/emnlp'+str(j)+'.txt'
    #print(j)
    f1=open(n1,'r')
    #f2=open(n2,'r')
    #f3=open(n3,'r')
    try:
        soup1=BeautifulSoup(f1,'html.parser')
        xn=j
        x.append(xn)
    except:
        continue
    #soup2=BeautifulSoup(f2,'html.parser')
    #soup3=BeautifulSoup(f3,'html.parser')
    result1=soup1.find_all('paper')
    #result2=soup2.find_all('paper')
    #result3=soup3.find_all('paper')
    result= len(result1)
    y.append(result)
```

```python


for j in range(0,19):
    n1='acl/acl'+str(j)+'.txt'
    #n2='naacl/naacl'+str(j)+'.txt'
    #n3='emnlp/emnlp'+str(j)+'.txt'
    #print(j)
    f1=open(n1,'r')
    #f2=open(n2,'r')
    #f3=open(n3,'r')
    try:
        soup1=BeautifulSoup(f1,'html.parser')
        xn=2000+j
        x.append(xn)
    except:
        continue
    #soup2=BeautifulSoup(f2,'html.parser')
    #soup3=BeautifulSoup(f3,'html.parser')
    result1=soup1.find_all('paper')
    #result2=soup2.find_all('paper')
    #result3=soup3.find_all('paper')
    result= len(result1)
    y.append(result)
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
#objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
objects=x
y_pos = np.arange(len(objects))
performance = y
plt.bar(y_pos, performance, align='center', alpha=1)
plt.xticks(y_pos, objects)
plt.ylabel('papers')
plt.xlabel('year')
plt.title('ACL paper publications')
 
plt.show()
```
