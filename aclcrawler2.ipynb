```python
import requests
r= requests.get('https://aclanthology.coli.uni-saarland.de/events/acl-2018')
```

```python
#print 100 characters
#print(r.text)
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(r.text, 'html.parser')
```

```python
results = soup.find_all('p')
```

```python
print(len(results))
```

```python
file = open("acl2018_long.txt","w")
```

```python
for i in range(1,257):
    x=1000+i
    fl='<Paper id ='+"\"P18-"+str(x)+"\">"
    
    paper=results[i].find('strong').text
    sl = "<title>" +str(paper)+"</title>"
    tl="<Venue> ACL </venue>"
    ttl="<type>long</type>"
    ltl="<Year>2018</Year>"
    ll= "</Paper>"
    ao="<Authors>"
    ac="</Authors>"
    aao="<Author>"
    aac="</Author>"
    print(fl)
    file.write(fl+'\n')
    print(sl)
    file.write(sl+'\n')
    print(tl)
    file.write(tl+'\n')
    print(ao)
    file.write(ao+'\n')
    
    
    for j in range(5,20):
        try:
            y=results[i].find_all('a')[j].text
            d=y.split(" ")
            print(aao)
            file.write(aao+'\n')
            z1="<FirstName>"+d[0]+"</FirstName>"
            z2="<SecondName>"+d[1]+"</SecondName>"
            print(z1)
            file.write(z1+'\n')
            print(z2)
            file.write(z2+'\n')
            print(aac)
            file.write(aac+'\n')
        except  :
            continue
    
    
    
    print(ac)
    file.write(ac+'\n')
    print(ttl)
    file.write(ttl+'\n')
    print(ltl)
    file.write(ltl+'\n')
    print(ll)
    file.write(ll+'\n')
    print('\n\n\n')
    file.write('\n\n')

file.close()


   
    
   
    
    
```
